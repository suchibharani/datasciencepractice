```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "http://simplilearn.com"
```

```python
result = requests.get(url)
page_content = result.content
soup_content = BeautifulSoup(page_content,'html.parser')
```

```python
soup_content
```

```python
print soup_content.prettify()
```

```python
soup_content.original_encoding
```

```python
soup_content.body.a.prettify(formatter='xml')
```

```python
def uppserCaseFn(strTxt):
    return strTxt.upper()

soup_content.body.a.prettify(formatter=uppserCaseFn)
```

```python
import pandas as pd
import string
from pprint import pprint
from time import time
```

```python
df_spam_collection = pd.read_csv('/Users/suchitra/Desktop/work/private/course/Data science/datasciencepractice/dataset/spam.csv', sep='\t',names=['response','message'])
```

```python
df_spam_collection.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
# from sklearn.grid_search import GridSearchCV - diff for v3 py
from sklearn.model_selection import GridSearchCV  
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('vect',CountVectorizer()),
    ('tfidf',TfidfTransformer()),
    ('clf',SGDClassifier())
])
```

```python
parameters = {'tfidf_use_idf':(True,False)}
```

```python
grid_search = GridSearchCV(pipeline,parameters,n_jobs=-1,verbose=1)
# print grid_search
print 'performing grid search now..'
pprint(parameters)
t0 = time()
grid_search.fit(df_spam_collection['message'],df_spam_collection['response'])
# print ('done in %0.3fs'% (time()-t0))
print()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
temp_data = [79,81,83,77,95,90]
wind_data = [14,11,10,13,9,12]
time_hrs = [1,2,3,4,5,6]
humidity_data = [73,76,81,81,76,79]
participation_data = [26,42,69,48,11,16]
```

```python
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.title('Temp')
plt.plot(time_hrs,temp_data,color= 'b',linestyle='-',linewidth=1)
plt.subplot(1,2,2)
plt.title('wind')
plt.plot(time_hrs,wind_data,color= 'r',linestyle='-',linewidth=1)

```

```python
plt.figure(figsize=(6,6))
plt.subplots_adjust(hspace=.25)
plt.subplot(2,1,1)
plt.title('Humidity')
plt.plot(time_hrs,humidity_data,color= 'b',linestyle='-',linewidth=1)
plt.figure(figsize=(6,6))
plt.subplots_adjust(hspace=.25)
plt.subplot(2,1,1)
plt.title('Precipitations')
plt.plot(time_hrs,participation_data,color= 'r',linestyle='-',linewidth=1)
```

```python
plt.figure(figsize=(9,9))
plt.subplots_adjust(hspace=.3)
plt.subplot(2,2,1)
plt.title('Temp (F)')
plt.plot(time_hrs,temp_data,color= 'b',linestyle='-',linewidth=1)
plt.subplot(2,2,2)
plt.title('Wind')
plt.plot(time_hrs,wind_data,color= 'r',linestyle='-',linewidth=1)
plt.subplot(2,2,3)
plt.title('Humidity')
plt.plot(time_hrs,humidity_data,color= 'g',linestyle='-',linewidth=1)
plt.subplot(2,2,4)
plt.title('Precipitation')
plt.plot(time_hrs,participation_data,color= 'y',linestyle='-',linewidth=1)
```

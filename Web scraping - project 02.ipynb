```python
from bs4 import BeautifulSoup
import urllib2
```

```python
url = "http://www.simplilearn.com/resources"
```

```python
webpage = urllib2.urlopen(url)
```

```python
sl_soup = BeautifulSoup(webpage,'html.parser')
```

```python
webpage.close()
```

```python
sl_soup.contents
```

```python
print sl_soup.prettify()
```

```python
print sl_soup.title
```

```python
for href in sl_soup.findAll('a',href=True):
    print href['href']
```
